SiO2 is normally obtained by oxidizing Si throught this chemical reaction \begin{equation} Si (s) + O_2 (g) = SiO_2 (s) \end{equation}

For wet oxydation the SiO2 is obtained by this reaction \begin{equation} Si (s) + H_2O (l) = SiO_2 + 2H_2 (g) \end{equation}
However for the reaction to take place you require a very high temperature between 800°C and 1100°C.

As you can observe the growth both happends by increaing the volume of the structure and throught diffusion in the Si

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# some handy functions to use along widgets
from IPython.display import display, Markdown, clear_output
# widget packages
import ipywidgets as widgets
from IPython.display import Image
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


#NEED TO TUNE THE PARAMETERS
# Constants
B = 1e-6  # linear rate constant (m/s)
k = 5e-12  # parabolic rate constant (m^2/s)
time_scale_factor = 100

# Time range (in seconds)
t = np.linspace(0, 10000, 100)

# Calculate oxide thickness using the Deal-Grove model
x = (-B + np.sqrt(B**2 + 4*k*t)) / 2

fig, ax = plt.subplots()

# Set up the plot elements
silicon = plt.Rectangle((0, 0), 1, 1, color='darkblue', label='Silicon')
oxide = plt.Rectangle((0, 1), 1, 0, color='orange', label='SiO2')
ax.add_patch(silicon)
ax.add_patch(oxide)

# Set up the axes and labels
ax.set_xlim(0, 1)
ax.set_ylim(0, 4)
ax.set_xticks([])
ax.set_yticks([0, 0.5, 1, 1.5 ,2])
ax.set_yticklabels([0, "500 nm", "1000 nm", "1500nm", "2000 nm"])
ax.set_ylabel("Thickness (µm)")
ax.legend()

def update(frame):
    oxide_thickness = x[frame] * 1e2  # Convert thickness from meters to micrometers
    oxide.set_height(oxide_thickness)
    ax.set_ylim(0, 2 + oxide_thickness)

ani = FuncAnimation(fig, update, frames=len(t), interval=100, repeat=False)

# Display the animation in JupyterLab
HTML(ani.to_jshtml())
```
